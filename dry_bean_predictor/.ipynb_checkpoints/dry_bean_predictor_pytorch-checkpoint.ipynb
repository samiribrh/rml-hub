# Dry Bean Classification Using Neural Networks (PyTorch)

## Overview

The goal of this task is to classify the type of dry bean based on multiple shape and texture features. We will explore the data, preprocess it, and create a neural network to predict the bean class.

## 1. Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

import torch
from torch import nn
import torch.nn.functional as F
from torch import optim
from torch.utils.data import TensorDataset, DataLoader
```

## 2. Loading and Exploring the Dataset

```python
# Load the dataset
df = pd.read_csv('Lecture_10_Dry_Bean_Dataset.csv')

# Show the first few rows
df.head()

# Show info about data types and missing values
df.info()

# Check for missing values
df.isnull().sum()

# Show summary statistics
df.describe()
```

### Dataset summary

The dataset has 13,611 samples with 17 columns:
16 numeric features, such as Area, Perimeter, MajorAxisLength, and others.

There are no missing values in any columns, so data cleaning related to missing data is not required.

<br>

```python
df['Class'].value_counts()
```

<br><br>
The target variable Class is categorical, containing seven distinct types of beans.

### Visualizing the class distribution

Before moving on, we will visualize the distribution of the target classes to better understand the balance of the dataset.

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Class', order=df['Class'].value_counts().index)
plt.title('Distribution of Dry Bean Classes')
plt.xlabel('Bean Class')
plt.ylabel('Number of Samples')
plt.xticks(rotation=45)
plt.show()
```

<br>The plot confirms the dataset is imbalanced. The classes DERMASON and SIRA have the largest sample sizes, while BOMBAY is the smallest class. This imbalance might affect the model’s ability to learn and generalize equally well across all classes.

### Visualizing numerical feature distribution

We'll visualize the distributions of all numerical features to get a picture before deciding on any feature selection.

```python
numeric_features = df.columns.drop('Class')

plt.figure(figsize=(16, 20))
for i, feature in enumerate(numeric_features, 1):
    plt.subplot(6, 3, i)
    sns.histplot(df[feature], bins=30, kde=True)
    plt.title(f'Distribution of {feature}')
plt.tight_layout()
plt.show()
```

### Feature Distributions

The histograms above show the distribution of all 16 numeric features in the dataset.

For example:

- Features like **Area**, **Perimeter**, **MajorAxisLength**, and **ConvexArea** have right-skewed distributions with a long tail towards higher values, indicating variability in bean sizes.
- Features such as **Solidity** and **ShapeFactor4** are tightly clustered near their maximum values, showing limited variance.
- Several features show multimodal or irregular distributions, which might indicate subgroups or complex relationships.

These observations suggest that **feature scaling** will be important before training the model to ensure numerical stability and better convergence. Additionally, some features may require transformations or careful consideration during modeling.


## 3. Preprocessing: Encoding and Scaling

```python
X = df.drop(columns=['Class'])
y = df['Class']

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Features shape:", X_scaled.shape)
print("Encoded target shape:", y_encoded.shape)
print("Sample scaled features:\n", X_scaled[:5])
print("Sample encoded targets:", y_encoded[:5])
```

<br> Features were scaled to standardize their ranges, which helps the model learn better. The target classes were converted to numbers so they can be used by the neural network.

The sample outputs confirm the scaling and encoding worked as expected.

## 4. Train-Test Split

We will now split the dataset into training and testing parts. Using an 80-20 split, the training set will be used to teach the neural network, while the test set will evaluate how well the model generalizes to unseen data. We'll keep the class distribution consistent in both sets by using stratified splitting.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

print("Training features shape:", X_train.shape)
print("Test features shape:", X_test.shape)
print("Training labels shape:", y_train.shape)
print("Test labels shape:", y_test.shape)
```

The data has been divided into training and testing parts, with roughly 80% used for training and 20% for testing. 

## 5. Converting Data to PyTorch Tensors

Next, we convert the training and testing data into PyTorch tensors. This step is essential because PyTorch models require inputs as tensors to perform efficient computations.


```python
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.long)
y_test_tensor = torch.tensor(y_test, dtype=torch.long)

print("X_train tensor shape:", X_train_tensor.shape)
print("y_train tensor shape:", y_train_tensor.shape)
print("X_test tensor shape:", X_test_tensor.shape)
print("y_test tensor shape:", y_test_tensor.shape)
print("X_train tensor type:", X_train_tensor.dtype)
print("y_train tensor type:", y_train_tensor.dtype)
```

The data has been successfully converted into PyTorch tensors with appropriate data types.

The tensor shapes confirm that the number of samples and features remain consistent after conversion. This means the data is now ready to be fed into the PyTorch neural network model.


## 6. Building the Feed-Forward Neural Network (PyTorch)

Now we will create a simple feed-forward neural network to classify the dry beans. The network will take the 16 input features and pass them through one or more hidden layers. The final layer will output probabilities for each bean class.

We will start with a baseline architecture of two hidden layers, and later experiment with different configurations.


```python
class BeanClassifier(nn.Module):
    def __init__(self, input_dim=16, hidden_dim1=64, hidden_dim2=32, output_dim=7):
        super(BeanClassifier, self).__init__()
        self.fc1 = nn.Linear(input_dim, hidden_dim1)
        self.fc2 = nn.Linear(hidden_dim1, hidden_dim2)
        self.fc3 = nn.Linear(hidden_dim2, output_dim)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)  # Output layer without activation for logits
        return x

model = BeanClassifier()

print(model)
```

The model consists of three fully connected layers. The first layer takes the 16 input features and outputs 64 values. The second layer reduces the dimension from 64 to 32. The final layer maps these 32 features to 7 outputs, corresponding to the seven bean classes.

## 7. Setting Up the Loss Function, Optimizer, and Training Loop

To train the neural network, we first define a suitable loss function and an optimizer. For multi-class classification, CrossEntropyLoss is appropriate.

We will use the Adam optimizer with a learning rate of 0.001.

Then, we set up a training loop that runs for a specified number of epochs, where in each epoch the model learns from the training data by minimizing the loss.


```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

epochs = 30
batch_size = 64

train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

# Initializing a list to store loss values
baseline_loss_history = []

for epoch in range(epochs):
    model.train()
    running_loss = 0.0
    for inputs, labels in train_loader:
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item() * inputs.size(0)

    epoch_loss = running_loss / len(train_loader.dataset)
    baseline_loss_history.append(epoch_loss)
    print(f"Epoch {epoch+1}/{epochs} - Loss: {epoch_loss:.4f}")
```

The training loss decreased significantly in the first few epochs, indicating that the model quickly learned useful patterns in the data. After around 10 epochs, the loss started to stabilize, showing gradual but consistent improvements.

By the end of 30 epochs, the loss is quite low, suggesting the model fits the training data well. However, to understand if the model generalizes properly, we need to evaluate its performance on the test set next.


## 8. Evaluating the Model on the Test Set

To assess how well the model generalizes to new data, we evaluate its performance on the test set. We will calculate the accuracy by comparing the predicted classes to the true labels. This will help us understand if the model has learned meaningful patterns or is overfitting to the training data.


```python
def evaluate_model(model, X_tensor, y_tensor):
    model.eval()
    with torch.no_grad():
        outputs = model(X_tensor)
        _, predicted = torch.max(outputs, 1)
    accuracy = accuracy_score(y_tensor.cpu(), predicted.cpu())
    return accuracy

train_accuracy = evaluate_model(model, X_train_tensor, y_train_tensor)
test_accuracy = evaluate_model(model, X_test_tensor, y_test_tensor)

print(f"Training accuracy: {train_accuracy:.4f}")
print(f"Test accuracy: {test_accuracy:.4f}")
```



**The model achieves a high accuracy of about 93.8% on the training data and 92.8% on the test data. This indicates that the neural network learned the patterns well and generalizes effectively to unseen data, with only a small gap between training and test accuracy.**

Such performance suggests the model is neither underfitting nor significantly overfitting, which is a strong result for this classification task.

## 9. Experimenting with Different Hyperparameters

To potentially improve the model’s performance, we can try changing the number of hidden layers and nodes, adjusting the learning rate, or using different optimizers like SGD or RMSprop. These changes help us understand how sensitive the model is to these factors and can lead to better accuracy or faster training.

We will start by increasing the number of hidden layers and nodes, and observe the effect on training and test accuracy.


```python
class DeeperBeanClassifier(nn.Module):
    def __init__(self, input_dim=16, hidden_dims=[128, 64, 32], output_dim=7):
        super(DeeperBeanClassifier, self).__init__()
        self.fc1 = nn.Linear(input_dim, hidden_dims[0])
        self.fc2 = nn.Linear(hidden_dims[0], hidden_dims[1])
        self.fc3 = nn.Linear(hidden_dims[1], hidden_dims[2])
        self.fc4 = nn.Linear(hidden_dims[2], output_dim)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        x = self.fc4(x)
        return x

deeper_model = DeeperBeanClassifier()

optimizer = optim.Adam(deeper_model.parameters(), lr=0.001)
criterion = nn.CrossEntropyLoss()
```

```python
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)

epochs = 30

deeper_loss_history = []

for epoch in range(epochs):
    deeper_model.train()
    running_loss = 0.0
    for inputs, labels in train_loader:
        optimizer.zero_grad()
        outputs = deeper_model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item() * inputs.size(0)

    epoch_loss = running_loss / len(train_loader.dataset)
    deeper_loss_history.append(epoch_loss)
    print(f"Epoch {epoch+1}/{epochs} - Loss: {epoch_loss:.4f}")
```

The deeper model shows a steady decrease in training loss over 30 epochs, starting from around 0.73 and ending near 0.17. This indicates that the additional layer and increased number of nodes allow the model to better fit the training data.

The loss is slightly lower than the simpler model’s final loss, suggesting the deeper architecture may capture more complex patterns. However, the true test will be how it performs on unseen data, so evaluating on the test set is the next step.

## 10. Evaluating the Deeper Model on Training and Test Sets

Now we will evaluate the deeper neural network’s performance on both training and test data, to see if the added complexity improves accuracy without causing overfitting.

```python
train_accuracy_deeper = evaluate_model(deeper_model, X_train_tensor, y_train_tensor)
test_accuracy_deeper = evaluate_model(deeper_model, X_test_tensor, y_test_tensor)

print(f"Training accuracy: {train_accuracy_deeper:.4f}")
print(f"Test accuracy: {test_accuracy_deeper:.4f}")
```

**The deeper model achieves about 93.9% accuracy on the training set and 92.7% on the test set. Compared to the previous simpler model, the changes are quite slight.**

The difference is not significant between models. It’s definitely not a win over the simpler model.

Next, we could try adjusting the learning rate or testing different optimizers to see if that helps.

## 11. Experimenting with Different Learning Rates

We will try lowering the learning rate to 0.0005 to see if it improves training stability and test accuracy.

Additionally, we will test the SGD optimizer.

We will train the model again with these new settings and compare the results.


```python
model_sgd = BeanClassifier()

criterion = nn.CrossEntropyLoss()
optimizer_sgd = optim.SGD(model_sgd.parameters(), lr=0.0005, momentum=0.9)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)

epochs = 30

sgd_loss_history = []

for epoch in range(epochs):
    model_sgd.train()
    running_loss = 0.0
    for inputs, labels in train_loader:
        optimizer_sgd.zero_grad()
        outputs = model_sgd(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer_sgd.step()
        running_loss += loss.item() * inputs.size(0)

    epoch_loss = running_loss / len(train_loader.dataset)
    sgd_loss_history.append(epoch_loss)
    print(f"Epoch {epoch+1}/{epochs} - Loss: {epoch_loss:.4f}")
```

With the SGD optimizer and a lower learning rate, the training loss started higher but steadily decreased throughout the epochs, showing consistent learning progress. 

Next, we should evaluate this model’s accuracy on training and test sets to see how this change affected generalization.

## 12. Evaluating the Model Trained with SGD Optimizer

We will check the accuracy of the model trained with the SGD optimizer on both training and test data, to compare its performance with the previous models.

```python
train_accuracy_sgd = evaluate_model(model_sgd, X_train_tensor, y_train_tensor)
test_accuracy_sgd = evaluate_model(model_sgd, X_test_tensor, y_test_tensor)

print(f"Training accuracy (SGD): {train_accuracy_sgd:.4f}")
print(f"Test accuracy (SGD): {test_accuracy_sgd:.4f}")
```

**The model trained with the SGD optimizer achieved 91.7% accuracy on the training set and 91.3% on the test set.**

Compared to the Adam optimizer results, this model shows slightly lower accuracy but better balance between training and test accuracy, indicating potentially less overfitting.

## Conclusion and Comparison of PyTorch Models

We tried a baseline model, a deeper model, and different optimizers. Plotting their training losses and accuracies side-by-side will help us clearly see the impact of these changes and make informed decisions.

Let's create plots to compare training loss curves and final accuracies on train and test sets.


```python
epochs_range = range(1, epochs + 1)

# Plot training loss curves
plt.figure(figsize=(10, 5))
plt.plot(epochs_range, baseline_loss_history, label='Baseline Model')
plt.plot(epochs_range, deeper_loss_history, label='Deeper Model')
plt.plot(epochs_range, sgd_loss_history, label='SGD Optimizer Model')
plt.xlabel('Epoch')
plt.ylabel('Training Loss')
plt.title('Training Loss Comparison')
plt.legend()
plt.grid(True)
plt.show()

# Plot accuracies in a bar chart
models = ['Baseline', 'Deeper', 'SGD']
train_accuracies = [train_accuracy, train_accuracy_deeper, train_accuracy_sgd]
test_accuracies = [test_accuracy, test_accuracy_deeper, test_accuracy_sgd]

x = np.arange(len(models))
width = 0.35

plt.figure(figsize=(8, 5))
plt.bar(x - width/2, train_accuracies, width, label='Train Accuracy')
plt.bar(x + width/2, test_accuracies, width, label='Test Accuracy')
plt.ylim(0.85, 1.0)
plt.ylabel('Accuracy')
plt.title('Training and Test Accuracy Comparison')
plt.xticks(x, models)
plt.legend()
plt.grid(axis='y')
plt.show()
```

The plots clearly show how the three models compare:

The training loss graph reveals that both the baseline and deeper models converge quickly and reach low loss values, indicating effective learning. The SGD optimizer model starts with a much higher loss but decreases steadily over time, showing a slower but consistent learning process.

The accuracy bar chart shows that the baseline and deeper models achieve similar training and test accuracies, slightly outperforming the SGD model. The small gap between training and test accuracy across all models suggests good generalization without severe overfitting.

These visualizations confirm that the baseline model is already quite strong, the deeper model doesn’t bring significant gains, and SGD, while stable, trades a bit of accuracy for smoother convergence.
